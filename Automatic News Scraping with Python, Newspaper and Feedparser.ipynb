```python
import  lxml
import newspaper
import feedparser
```

```python
def scrap_news_from_feed(feed_url):
    articles = []
    feed = feedparser.parse(feed_url)
    for entry in feed.entries:
        # create a newpaper article object
        article = newspaper.Article(entry.link)
        # Download and parse the article
        article.download()
        article.parse()

        # extract relevant information from the article  like
        # title,authot,date of publish , content
        articles.append({
            'Title':article.title,
            'Author':article.authors,
            'Publish date':article.publish_date,
            'Content':article.text
            
        })
    return articles

feed_url = 'http://feeds.bbci.co.uk/news/rss.xml'
articles = scrap_news_from_feed(feed_url)    

# print the extracted imformation
for article in articles:
    print("Title:", article['Title'])
    print("Author:",article['Author'])
    print("Publish date:",article['Publish date'])
    print("Content:",article['Content'])
    print()


```
